<H3>Data wrangling ML features => Ozone, CO, PM2.5, & N02</H3>
Focus of this notebook is to import BigQuery results and put into dataframe that minimizes duplicate "key" values identified in previous iterations of the project work. This notebook builds of code from project team members.

<hr>
<h4>1. Loading the data</h4>

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
# file = "env_data.json"
file = "bigquery.csv"

# Read CSV
# env_data = pd.read_json(file, dtype=False, lines=True)
# env_data.head()
query_dF = pd.read_csv(file)
query_dF.head(2)
```

<h4>2. Exploring the dataframe</h4>

```python
#Setup back-up copy of dataframe & simple 'df' naming convention
query_dF_copy = query_dF
df = query_dF
df.shape
df.columns
```

```python
#Exploring the dataframe for columns potentially generating duplicates 
df['year'].unique()
df['pollutant_standard'].unique()
df['parameter_code'].unique()
df['parameter_name'].unique()
df.groupby(['parameter_name','parameter_code']).max().head(4)
```

<hr>
<h4>2b. ... "Houston...we have a problem" with duplicates</h4>
Replicating Ed's "check for duplicates - seems like we have multiple values for the same year, state, county, and parameter"

```python
# Harris County, TX fips 48201 (:-) Houston is located in Harris County
test_48201=df.loc[(df['fips']==48201) &
                  (df['year']==2014) &
                  (df['parameter_code']==44201)
                 ]
test_48201.to_csv('test_48201.csv')
test_48201.shape
test_48201.head(4)

# Autauga, AL fips 1001
test_1001=df.loc[(df['fips']==1001)
                 & (df['year']==1990)
                 & (df['parameter_code']==44201)
                ]
test_1001.to_csv('test_1001.csv')
test_1001.shape
test_1001.head(4)
```

<hr>
<h4>3. Transforming/cleaning the dataframe to minimize duplicates</h4>

```python
# Multiple pollutant_standard for same measurement generates duplicates
df['pollutant_standard'].unique()

# From list above, selecting standards thought to cover more years
select_pollutant_standards = df.loc[(df['pollutant_standard']=="Ozone 8-Hour 1997") |
                         (df['pollutant_standard']=="CO 8-hour 1971") |
                         (df['pollutant_standard']=="NO2 Annual 1971") |
                         (df['pollutant_standard']=="PM25 24-hour 2006")                   
                        ]
# Viewing  & verifying selection via groupby
select_pollutant_standards.groupby(['parameter_name','parameter_code','pollutant_standard']).max().head(4)
```

```python
#Checking what the de-duplication effort yielded
select_pollutant_standards.shape
df.shape
print(f'Before ' + str(len(df)) + ' rows and after '+ str(len(select_pollutant_standards)) + ' rows')
```

<hr>
<h4>3b. ... "Houston...we still have a problem" with duplicates related to <u>monitoring sites</u></h4>
Is it legit to keeps monitoring sites duplicates and then aggregate, as suggested?

```python
select_df = select_pollutant_standards
# Harris County, TX fips 48201 (:-) Houston is located in Harris County
test2_48201=select_df.loc[(select_df['fips']==48201) &
                  (select_df['year']==2014) &
                  (select_df['parameter_code']==44201)
                 ]
test2_48201.to_csv('test2_48201.csv')
test2_48201.shape
test2_48201.head(2)
test2_48201['site_num'].unique()
print(f'Duplicates from Harris County, Texas (fips 48201)')
print(f' =>Before ' + str(len(test_48201)) + ' rows and after '+ str(len(test2_48201)) + ' rows')
print(f' =>...duplicates that remain after are from ' + \
      str(len(test2_48201['site_num'].unique())) + ' monitoring sites for Ozone alone!')
```

<hr>
<h4>4. Transpose (pivot_table) the pollutant parameters to setup df for merging with the asthma frame</u></h4>
Assuming it is legit to keep monitoring sites duplicates and then aggregate, as suggested?

```python
select_df.columns
```

```python
columns_to_index = ['year','county_name','fips','site_num','latitude','longitude','state_name']
```

```python
transposed=select_df.pivot_table(index = columns_to_index,
            columns = ['parameter_name'],fill_value=0).reset_index()
transposed.shape
```

```python
# Update columms names by combining measure name
transposed.columns = transposed.columns.map('|'.join).str.strip('|')
transposed.head()
```

```python
# Harris County, TX fips 48201; Houston is located in Harris County
test3_48201=transposed.loc[(transposed['year']==2014)&
                          (transposed['fips']==48201)
                         ]
test3_48201.to_csv('test3_48201.csv')
test3_48201.shape
test3_48201.head()
test3_48201['site_num'].unique()
len(test3_48201['site_num'].unique())

# Autauga, AL fips 1001
test3_1001=transposed.loc[(transposed['year']==1990)&
                          (transposed['fips']==1001)
                         ]
test3_1001.to_csv('test3_1001.csv')
test3_1001.shape
test3_1001.head()
test3_1001['site_num'].unique()
len(test3_1001['site_num'].unique())
```
